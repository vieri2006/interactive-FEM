```python
from FEM import *
```

## Features:
1. Create object by `FEM(params)`, look at `help(FEM)` for more details )
2. `FEM_object.changeNode(node_index, value)` to change node
3. `FEM_object.changeK(strings_of_func)` to change conductivity as a function of x (disclaimer: belum di verifikasi krn tidak menemukan solusi analitik jika K fungsi dari x)
4. `FEM_object.changeQ(strings_of_func)` to change heat generation as a function of x
5. `plotT(FEM_object)` to plot Temperature distribution
6. `plotShapeFunction(FEM_object)` to plot Shape Function used
7. `plotTExpansion(FEM_object)` to plot Temperature Weighted Shape Function

## Limitation:
1. Max number of elements for {orde 1 = 4; orde 2 = 6; orde 3 = 9}, otherwise the computation will took very long
2. K variation is not yet verified with the analytical eq!

## DEMO: Basic
P.S. Try the cubic and linear yourself :)

```python
fem_obj = FEM(orde = 2, Nelements = 4, Lnum = 1, T_Lnum = 10, q0num = 10, knum = 4, Qnum = 1)
```

```python
fem_obj.evaluate()
```

```python
fem_obj.sols_qL, fem_obj.sols_T
```

```python
plotT(fem_obj)
```

```python
plotShapeFunction(fem_obj)
```

```python
plotTExpansion(fem_obj)
```
